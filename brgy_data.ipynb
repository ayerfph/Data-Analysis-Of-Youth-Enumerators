Data Analysis Of Youth Profiled 

```python
# Import libraries
import pandas as pd
from itables import show
```

```python
# Import raw data and convert into dataframe
raw_data = pd.read_csv('kabataankonek_cleaned_raw.csv')

# Check if dataframe was created
raw_data.head()
```

```python
# Using itables for better view and quick search filtering tables
show(raw_data)
```

Total Days Youth Profiled Per Barangay

```python
# Creating dataframe for getting data of total days youth profiled per barangay
days_per_barangay = pd.DataFrame()

# Creating columns
days_per_barangay['barangay'] = raw_data['barangay'].unique()

days = [f'day_{i}' for i in range(1, 22)]
days_per_barangay = raw_data.groupby('barangay')[days].sum().reset_index()

# Display the table
show(days_per_barangay)
```

```python
# Display result table
days_per_barangay
```

Total Weeks Youth Profiled Per Barangay

```python
# Creating dataframe for getting data of total weeks youth profiled per barangay
weeks_per_barangay = pd.DataFrame()

# Creating columns
weeks_per_barangay['barangay'] = days_per_barangay['barangay']
weeks_per_barangay['week_1'] = days_per_barangay[['day_1', 'day_2', 'day_3']].sum(axis=1)
weeks_per_barangay['week_2'] = days_per_barangay[['day_4', 'day_5', 'day_6']].sum(axis=1)
weeks_per_barangay['week_3'] = days_per_barangay[['day_7', 'day_8', 'day_9']].sum(axis=1)
weeks_per_barangay['week_4'] = days_per_barangay[['day_10', 'day_11', 'day_12']].sum(axis=1)
weeks_per_barangay['week_5'] = days_per_barangay[['day_13', 'day_14', 'day_15']].sum(axis=1)
weeks_per_barangay['week_6'] = days_per_barangay[['day_16', 'day_17', 'day_18']].sum(axis=1)
weeks_per_barangay['week_7'] = days_per_barangay[['day_19', 'day_20', 'day_21']].sum(axis=1)

# Display the table
show(weeks_per_barangay)
```

```python
# Display result table
weeks_per_barangay
```

Total Months Youth Profiled Per Barangay

```python
# Creating dataframe for getting data of total months youth profiled per barangay
months_per_barangay = pd.DataFrame()

# Creating columns
months_per_barangay['barangay'] = weeks_per_barangay['barangay']
months_per_barangay['september'] = weeks_per_barangay[['week_1', 'week_2', 'week_3', 'week_4']].sum(axis=1)
months_per_barangay['october'] = weeks_per_barangay[['week_5', 'week_6', 'week_7']].sum(axis=1)

# Display the table
show(months_per_barangay)
```

```python
# Display result table
months_per_barangay
```

```python
# Creating dataframe for getting data of total months youth profiled per barangay
youth_profiled_per_barangay = pd.DataFrame()

youth_profiled_per_barangay['barangay'] = months_per_barangay['barangay']
youth_profiled_per_barangay['total'] = months_per_barangay[['september', 'october']].sum(axis=1)

show(youth_profiled_per_barangay)
```
