Analyze Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from itables import show
```

```python
df = pd.read_csv('kabataankonek_cleaned_raw.csv')

df.head(1)
```

Getting Sum of Days Per Barangay

```python
days = [f'day_{i}' for i in range(1, 22)]

sum_days_per_brgy = pd.DataFrame()
sum_days_per_brgy['barangay'] = df['barangay'].unique()

for day in days:
    sum_days_per_brgy[day] = df.groupby('barangay')[day].sum().values
```

```python
show(sum_days_per_brgy)
```

Getting Sum Of Weeks Per Barangay

```python
sum_weeks_per_brgy = pd.DataFrame()
sum_weeks_per_brgy['barangay'] = sum_days_per_brgy['barangay']

sum_weeks_per_brgy['week_1'] = sum_days_per_brgy[['day_1', 'day_2', 'day_3']].sum(axis=1)
```

```python
sum_weeks_per_brgy.head(1)
```

```python
sum_weeks_per_brgy['week_2'] = sum_days_per_brgy[['day_4', 'day_5', 'day_6']].sum(axis=1)
sum_weeks_per_brgy['week_3'] = sum_days_per_brgy[['day_7', 'day_8', 'day_9']].sum(axis=1)
sum_weeks_per_brgy['week_4'] = sum_days_per_brgy[['day_10', 'day_11', 'day_12']].sum(axis=1)
sum_weeks_per_brgy['week_5'] = sum_days_per_brgy[['day_13', 'day_14', 'day_15']].sum(axis=1)
sum_weeks_per_brgy['week_6'] = sum_days_per_brgy[['day_16', 'day_17', 'day_18']].sum(axis=1)
sum_weeks_per_brgy['week_7'] = sum_days_per_brgy[['day_19', 'day_20', 'day_21']].sum(axis=1)
```

```python
show(sum_weeks_per_brgy)
```

Getting Sum of Months Per Barangay

```python
sum_months_per_brgy = pd.DataFrame()
sum_months_per_brgy['barangay'] = sum_days_per_brgy['barangay']

sum_months_per_brgy['september'] = sum_weeks_per_brgy[['week_1', 'week_2', 'week_3', 'week_4']].sum(axis=1)
sum_months_per_brgy['october'] = sum_weeks_per_brgy[['week_5', 'week_6', 'week_7']].sum(axis=1)
```

```python
show(sum_months_per_brgy)
```

Getting The Total Youth Profiled Per Barangay

```python
total_per_brgy = pd.DataFrame()
total_per_brgy['barangay'] = sum_days_per_brgy['barangay']

total_per_brgy['total_youth_profiled'] = sum_months_per_brgy[['september','october']].sum(axis=1)
```

```python
show(total_per_brgy)
```
